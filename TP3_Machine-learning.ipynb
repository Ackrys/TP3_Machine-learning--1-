### Affichage du nombre de films de chaque genre

Certains films sont associés à plusieurs genres : ils devront donc être comptabilisés dans chacun des genres (un genre est représenté par un seul mot).

```python
def getCountByGenre() :
    result = {}

    for row in csv_data.itertuples() :
        str = row.Genre
        str = str.split(",")
        for i in range(len(str)) :
            if str[i] in result.keys() :
                result[str[i]] += 1
            else :
                result[str[i]] = 1
    return result

def getCountByGenre2(result) :

    for row in result :
        str = row.Genre
        str = str.split(",")
        for i in range(len(str)) :
            if str[i] in result.keys() :
                result[str[i]] += 1
            else :
                result[str[i]] = 1
    print(result)
    return result

print(getCountByGenre())
```

### Affichage du nombre de films par genre et par année

Afficher un histogramme donnant la répartition du nombre de films par genre, pour chaque année.

```python
import numpy as np
import matplotlib.pyplot as plt

result = {}

for row in csv_data.itertuples() :
    str = row.Year
    stg = row.Genre
    stg = stg.split(",")
    for i in range(len(stg)):
        if (str, stg[i]) in result.keys() :
            result[(str, stg[i])] += 1
        else :
            result[(str, stg[i])] = 1

print(result)

# Split the dataset by year
dataset_by_year = result.groupby('Year')

# For each year, get the number of moviers for each genre
count_by_genre_by_year = {}
for year, group in dataset_by_year:
    count_by_genre_by_year[year] = getCountByGenre2(group)

# Convert the dictionary to a dataframe
count_by_genre_by_year_df = pnd.DataFrame.from_dict(count_by_genre_by_year, orient='index')
count_by_genre_by_year_df.head()

# Plot the dataframe and place the legend at the right of the plot
count_by_genre_by_year_df.plot(kind='bar', stacked=True)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

### Création d'un dataframe contenant uniquement le titre du film, son résumé ainsi que son genre

```python
data = {'title': [],
        'resume': [],
        'genre': []}
i = 0
df = pnd.DataFrame(data)
for row in csv_data.itertuples() :
        t = row.Title
        r = row.Description
        g = row.Genre
        df = df.append({'title': t,
                        'resume' : r,
                        'genre': g}, ignore_index=True)
print(df)
```


## 2. Classification binaire d'un genre de film choisi

Vous devrez tout d'abord choisir un genre de films parmi ceux présents dans les données. L'objectif sera ensuite de créer un classifieur permettant d'indiquer si un film appartient au genre choisi, en utilisant la description du film.

### Ajout d'une colonne correspondant à l'appartenance ou non du film au genre choisi

Le genre choisi pour la suite est Action.

Si le film appartient à ce genre, il aura la valeur 1 dans cette nouvelle colonne et il aura la valeur 0 dans le cas contraire.

```python
choix = "Action"

belongsToIt = []

for row in df.itertuples() :
    g = row.genre
    g = g.split(",")
    tmp = -1
    for i in range(len(g)) :
        if g[i] == choix :
            tmp = 1
        else :
            if tmp != 1 :
                tmp = 0
    belongsToIt.append(tmp)


classif = df
classif['BelongsToTheGenre'] = belongsToIt
print(classif)
```
